```python
import os
from skimage import io
import numpy as np
import pandas
from tqdm import tqdm

from PIL import Image
Image.MAX_IMAGE_PIXELS = 1000000000

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
city_dir = "../../data/cowc_cities"

city_txts = os.listdir(city_dir)
print(city_txts)    
```

```python
data_root = "../../data/cowc/datasets/ground_truth_sets"

for city_txt in city_txts:
    city_name, _ = os.path.splitext(city_txt)
    
    with open(os.path.join(city_dir, city_txt)) as f:
        files = f.readlines()
    
    for file in files:
        file = file.rstrip()
        
        label_file = file + "_Annotated_Cars.png"
        label_path = os.path.join(data_root, os.path.join(city_name, label_file))
        
        print("Loading {} ...".format(label_file))
        
        label = io.imread(label_path)
        
        count = (label[:, :, 0] > 2).sum()
        
        print(label.shape, count)
        print()
```
